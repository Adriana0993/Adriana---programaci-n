# Pregunta 12

Para s√≥lo los pa√≠ses de Centroam√©rica ['El Salvador', 'Honduras', 'Guatemala', 'Costa Rica', 'Nicaragua'] obtener un reporte por a√±o de las emisiones totales en Centroam√©rica usando el dataframe 'df_emisions_co2'?
Se espera un dataframe de 2 columnas: 'Year' y 'total_por_a√±o'

```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta
centroamerica = ['El Salvador' , 'Honduras' , 'Guatemala' , 'Costa Rica', 'Nicaragua']
df_centroamerica = df_emisions_co2[df_emisions_co2['Country'].isin(centroamerica)].copy()
total = df_centroamerica.groupby(['Year' , 'Country'])['Value'].sum()
total_por_a√±o = total.unstack(level='Year')
print(f'Las emisiones totales por a√±o para Centramerica son: {total_por_a√±o}')
```

# Pregunta 13

Para s√≥lo los pa√≠ses de Centroam√©rica ['El Salvador', 'Honduras', 'Guatemala', 'Costa Rica', 'Nicaragua'] obtener un reporte por a√±o de las emisiones en promedio en Centroam√©rica usando el dataframe 'df_emisions_co2_per_capita'?
Se espera un dataframe de 2 columnas: 'Year' y 'promedio_por_a√±o'

```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta
df_centro = df_emisions_co2_per_capita[df_emisions_co2_per_capita['Country'].isin(centroamerica)].copy()
promedio_centroamerica = df_centroamerica.groupby(['Year' , 'Country'])['Value'].mean()
promedio_por_a√±o = promedio_centroamerica.unstack(level='Year')
print(f'El promedio de emisiones por a√±o para Centroamerica es: {promedio_por_a√±o} ')
```

# Pregunta 14

# Pregunta 15

Usa el dataframe  'df_emisions_co2_per_capita' con las columnas Year y promedio_por_a√±o para crear una gr√°fica de l√≠neas que muestre la evoluci√≥n del promedio de emisiones per c√°pita en Centroam√©rica a lo largo del tiempo; el eje X debe ser el a√±o y el eje Y el promedio de emisiones.

```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunt

import matplotlib.pyplot as plt
a√±o = df_centro['Year']
promedio_por_a√±o = df_centro['Value']

fig, ax = plt.subplots()
ax.plot(a√±o, promedio_por_a√±o , marker= 'o' , linestyle= '-' , color='b' , label='Promedio de emisiones')
ax.set_title('Evoluci√≥n del promedio de emisiones per c√°pita en  Centroam√©rica', fontsize=14)
ax.set_xlabel('a√±o')
ax.set_ylabel('Promeedio por a√±o(promedio_por_a√±o)')
ax.grid(True, linestyle='--' , alpha=0.7)
fig.tight_layout()
plt.show
```
